```python
def depth_plot_point_size(depth, min_depth, max_depth):
    '''
    Normalize depth to a size range 20 -> 150 for visualization
    '''
    normalized = (depth - min_depth) / (max_depth - min_depth)
    return (normalized * 130) + 20
```

```python
def resample_by_depth(group, num_depth_samples):
    """
    Given a group's data, pick up to num_samples points spaced along DEPTH.
    """
    if len(group) <= num_depth_samples:
        return group
    
    indices = np.linspace(0, len(group) - 1, num_depth_samples, dtype=int)
    return group.sort_values('DEPTH').iloc[indices]
```

```python
def create_depth_legend_handles(min_depth, max_depth, depth_plot_point_size):
    """
    Returns a list of handles for shallow, mid-depth, and deep points.
    """
    depth_categories = [
        (min_depth, "Shallow"),
        ((min_depth + max_depth) / 2, "Mid-depth"),
        (max_depth, "Deep")
    ]

    depth_handles = []
    for d, label in depth_categories:
        handle = plt.scatter(
            [], [], 
            s=depth_plot_point_size(d, min_depth, max_depth),
            color='gray',
            alpha=0.7,
            edgecolor='k',
            label=f"{label} ({d:.1f}m)"
        )
        depth_handles.append(handle)

    return depth_handles
```

```python
def create_station_legend_handles(station_ids, color_map):
    """
    Returns a list of handles (Line2D) for each station in station_ids.
    """
    station_handles = []
    for s in station_ids:
        handle = Line2D(
            [0], [0], 
            marker='o',
            color='w',
            markerfacecolor=color_map[s],
            markersize=8, 
            label=f"Station {s}"
        )
        station_handles.append(handle)
    return station_handles
```

```python
def plot_param_vs_time(csv_path, param, y_axis_label, num_depth_samples=10):
    df = pd.read_csv(csv_path)    
    df['STAT_ARRIV'] = pd.to_datetime(df['STAT_ARRIV'], errors='coerce')    
    
    df_filtered = df.dropna(subset=['STAT_ARRIV', 'DEPTH', f'{param}'])
    df_filtered['sample_date'] = (df_filtered['STAT_ARRIV'].dt.date)
    
    MIN_DEPTH = df_filtered['DEPTH'].min()
    MAX_DEPTH = df_filtered['DEPTH'].max()
    
    df_filtered['depth_point_size'] = df_filtered['DEPTH'].apply(depth_plot_point_size, args=(MIN_DEPTH, MAX_DEPTH))
    
    station_colors = plt.cm.tab10.colors
    station_ids = df_filtered['STAT_ID'].unique()
    color_map = {
        station: station_colors[i % len(station_colors)]
        for i, station in enumerate(station_ids)
    }
    fig, ax = plt.subplots(figsize=(14, 16))

    for (date, station), group in df_filtered.groupby(['sample_date', 'STAT_ID']):
        date = date.strftime('%m/%d/%Y')
        
        group = resample_by_depth(group, num_depth_samples)

        plot_time = group['STAT_ARRIV'].iloc[0]
        station_color = color_map[station]
            
        ax.scatter(
            [date] * len(group),
            group[f'{param}'],
            s=group['depth_point_size'],
            color=station_color,
            alpha=0.7,
            edgecolor='k',
            linewidth=0.5
        )

        ax.plot(
            [date] * len(group),
            group[f'{param}'],
            '-',
            color=station_color,
            linewidth=1,
            alpha=0.5
        )

        ax.set_title(f'{param} vs. Time', fontsize=14)
        ax.set_xlabel(f'{param}')
        ax.set_ylabel(f'{y_axis_label}', fontsize=12)
        ax.grid(True, alpha=0.3)
    
    depth_handles = create_depth_legend_handles(MIN_DEPTH, MAX_DEPTH, depth_plot_point_size)
    station_handles = create_station_legend_handles(station_ids, color_map)

    depth_legend = ax.legend(
        handles=depth_handles,
        loc='upper right',
        bbox_to_anchor=(1.0, 1.0),
        title="Depth"
    )

    station_legend = ax.legend(
        handles=station_handles,
        loc='upper right',
        bbox_to_anchor=(1.0, 0.90),
        title="Station ID"
    )

    ax.add_artist(depth_legend)

    plt.savefig(f'{param}_vs_time_and_depth.png', dpi=300, bbox_inches='tight')
    plt.show()
```
